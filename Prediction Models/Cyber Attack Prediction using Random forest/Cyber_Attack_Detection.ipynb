# Cyber Attack Detection
To enhance cybersecurity, machine learning algorithms can be utilized to detect cyber attacks. This approach involves analyzing network data to find patterns and correlations indicative of potential threats. Implementing machine learning improves the accuracy and efficiency of detecting attacks, thereby strengthening the security of digital networks and systems.

#Understanding Data Set
### The dataset categorizes nine types of cyber attacks:

- Fuzzers: Sending random data to test system resilience and find vulnerabilities.
- Analysis: Examining a system to identify weaknesses and potential targets.
- Backdoors: Creating hidden entry points for unauthorized future access.
- DoS (Denial of Service): Disrupting system functionality to make it unavailable to users.
- Exploits: Leveraging system vulnerabilities to gain unauthorized access or control.
- Generic: Includes various attack types not covered by other categories.
- Reconnaissance: Gathering information on a target system to plan future attacks.
- Shellcode: Executing malicious code, often shell scripts, on a target system.
- Worms: Self-replicating malware that spreads and causes harm.

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns = ['attack_cat', 'label'])
y = df['label'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=11)

feature_names = list(X.columns)

print("X_train shape: ", X_train.shape)
print("y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test shape: ", y_test.shape)
```

Decision Tree:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [2, 4],
    'min_samples_split': [2, 4],
    'min_samples_leaf': [1, 2]
}

dt = DecisionTreeClassifier()

grid_search = GridSearchCV(dt, param_grid, cv=5, scoring='recall')
grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best recall score:", grid_search.best_score_)
```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score

clf=grid_search.best_estimator_
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

recall = recall_score(y_test, y_pred)
print("Recall: ", recall)
```

```python
from sklearn.tree import export_text
import dtreeviz

print(":::::::> The RULES FOR HIGH RECALL RATE <::::::: \n" ,export_text(clf,feature_names=feature_names))

viz_model = dtreeviz.model(clf,
                           X_train=X_train, y_train=y_train,
                           feature_names=feature_names)
viz_model.view()
```

```python
X_test = X_test.reset_index(drop=True)

rules= "(sttl <= 61.00 & sinpkt<= 0.00) | (sttl >  61.00 )"

ind = X_test.query(rules).index

X_test_2 = X_test.loc[ind,:]
y_test_2 = y_test[ind]

print(X_test.shape)
print(X_test_2.shape)
print("filtered data" , (1- np.round(X_test_2.shape[0] / X_test.shape[0],2))*100, "%")
```

The rule-based system effectively filtered 23% of network traffic for further review. Next, we will use the trained model to detect cyber attacks within this filtered dat

```python
from sklearn.metrics import accuracy_score, precision_score
```

```python
def model_evaluation(model):
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test_2)

    accuracy = accuracy_score(y_test_2, y_pred)
    recall = recall_score(y_test_2, y_pred)
    precision = precision_score(y_test_2, y_pred)
    print("Recall: ", recall)
    print("Precision: ", precision)
    print("Accuracy: ", accuracy)

    cross = pd.crosstab(pd.Series(y_test_2, name='Actual'), pd.Series(y_pred, name='Predicted'))
    plt.figure(figsize=(5, 5))
    sns.heatmap(cross, annot=True,fmt='d', cmap="YlGnBu")
    plt.show()

    return {'Recall' : recall}
```

```python
results = {}
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=11)

results['Random Forest Model'] = model_evaluation(rf)
```

```python
comparision = pd.DataFrame(results)
comparision
```

The Wilcoxon rank sum test is a non-parametric test that does not assume normality in the data. It calculates a Z-statistic to measure variation between recall scores and a p-value to assess the probability of observing results as extreme as those found, assuming the null hypothesis (no difference in recall) is true. A p-value less than 0.05 indicates a statistically significant difference in recall among the models, allowing us to reject the null hypothesis

```python
from scipy.stats import wilcoxon

z_statistic, p_value = wilcoxon([comparision.iloc[0][0], comparision.iloc[0][1], comparision.iloc[0][2]])

# Print the results
print('Z-statistic:', z_statistic)
print('p-value:', p_value)

# Interpret the results
if p_value < 0.05:
    print('The difference in the recall of the three models is statistically significant.')
else:
    print('The difference in the recall of the three models is not statistically significant.')
```
